```python
import numpy as np
```

```python
def angle(a,b,c):
    BA = a - b
    BC = c - b
    
    
    dot_product = np.dot(BA, BC)
    
    
    x=np.linalg.norm(BA)
    y=np.linalg.norm(BC)
    
    
    cos_angle=dot_product/(x*y)
    
    angle_in_degrees=np.degrees(np.arccos(cos_angle))
    
    return angle_in_degrees
```

```python
N = []
c=0
with open('C:/Users/meenu.DESKTOP-FACTER5/Desktop/193l.pdb') as infile:
    for line in infile:
        if line[17:21]=="GLY " and line[:4] == 'ATOM' and (line[77:79]=='N ' or line[77:79]=='CA' or line[77:79]=='C ' or line[77:79]=='O '):
            try:
                pl=[float(line[31:39]), float(line[39:47]), float(line[47:55])]
                                        
                N.append(pl)
            except NameError:
                print("Error")
                
cd=np.array(N)
        

for i in range(1,len(cd),4):
    print('N-CA-C <',cd[i-1],'\n\t',cd[i],'\n\t',cd[i+1],'\n')
    x=angle(cd[i-1],cd[i],cd[i+1])
    print('Angle N-CA-C = ',x,'\n')
    
    for j in range(i+1,i+2):        
        print('CA-C-O <',cd[j-1],'\n\t',cd[j],'\n\t',cd[j+1],'\n\n')
        y=angle(cd[j-1],cd[j],cd[j+1])
        print('Angle CA-C-O = ',y,'\n')
        
        
```
